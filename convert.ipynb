# **CONVERT**

The original author dumped the flow velocity, pressure, and node coordinates at each time step in the raw file **fixed_cylinder_atRe100**. It is not in any standard format, but can be read using the `read_flow` method in `text_flow.py`. Using this, we load the simulation which includes the Reynold's number, node locations, velocity field and pressure field.

```python
# Dependencies,
import h5py
from text_flow import read_flow

# Reading raw data,
Re, _, times, nodes_X, nodes_Y, U, V, p = read_flow("fixed_cylinder_atRe100")
```

Now we save the data we need as a `HDF5` file (a standard open-soruce format),

```python
# Saving the velocity components, node locations and pressure field into a HDF5 file,
with h5py.File("cylinder_Re100.h5", "w") as file:
    file.attrs["Re"] = Re
    file.create_dataset("t", data=times)
    file.create_dataset("u", data=U)
    file.create_dataset("v", data=V)
    file.create_dataset('p', data=p)
    file.create_dataset('x', data=nodes_X)
    file.create_dataset('y', data=nodes_Y)
```
